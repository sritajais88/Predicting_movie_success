```python
# import 
import sys
from pathlib import Path
PROJECT_ROOT= Path().resolve().parent
sys.path.append(str(PROJECT_ROOT))


```

```python
import numpy as np
import pandas as pd
from src import config 
```

```python
## LOad data
DATA_FILE_PATH = config.DATA_PATH
CLEANED_DATA_FILE_PATH = config.CLEANED_DATA_Path
print("Raw data Path/n:",DATA_FILE_PATH)
print("Cleaned Data Path/n:",CLEANED_DATA_FILE_PATH)
```

```python
## Load Data
data =pd.read_csv(DATA_FILE_PATH)
data.head(10)
```

```python
# data backup
df= data.copy()
```

```python
# Meta data
print("No.of rows:",df.shape[0])
print("\n No.of columns:",df.shape[1])
print("\n All columns:",df.columns.to_list())
print ("\n Data-types:",df.dtypes)
print("\n Info:")
df.info()
print ("\n no.of Duplicates:",df.duplicated().sum())
```

```python
## Removing duplicates

df=df.drop_duplicates()
print("No.of duplicates finally:",df.duplicated().sum())
```

```python
df.describe()
```

```python
### Remove irrelevant columns


drop_columns=["director_name","actor_2_name","actor_1_name","movie_title","actor_3_name","plot_keywords","movie_imdb_link"]
df.drop(columns=drop_columns,axis=1,inplace =True)
```

```python
print ("Name of columns:",df.columns.to_list())
```

```python
## Save the data
df.to_csv(CLEANED_DATA_FILE_PATH,index= False)
print("Data is saved successfully")
```
